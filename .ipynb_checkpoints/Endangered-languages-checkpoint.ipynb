```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataf = pd.read_csv(r'C:\Users\david\Desktop\endangered-languages\data.csv')       # load the dataset
```

```python
dataf                       # show loaded dataset
```

```python
plt.figure(figsize=(12,6))                                 # set the size of the chart
dng = sns.countplot(data=dataf, x='Degree of endangerment')         # plot the bar chart, assigning to a variable for customization
dng.set_xticklabels(dng.get_xticklabels(), rotation=90);            # align x axis labels vertically for proper visibility
dng.set(title='Number of languages per degree of endangerment');
```

```python
EastA = dataf[dataf['Countries'].isin(['Kenya', 'Ethiopia', 'Tanzania', 'Uganda', 'Rwanda', 'Burundi', 'South Sudan', 'Somalia'])]   # creating a subset of the data containing only the required East African countries from the original dataset, and assigning to a new variable, EastA
```

```python
EastA                  # display the new dataset
```

```python
plt.figure(figsize=(12,6))                                       # set the size of the chart to be drawn
EA = sns.countplot(data=EastA, x='Degree of endangerment')                     # create chart of endangered languages in given East African countries, assign to a variable
EA.set_xticklabels(EA.get_xticklabels(), rotation=90);            # align x labels verically
EA.set(title='Number of languages per degree of endangerment in East Africa');
```

```python
spk = sns.catplot(data=dataf, x='Degree of endangerment', y='Number of speakers', kind='bar')           # chart of number of speakers under each degree of endangerment
spk.set(title='Number of Speakers per degree of endangerment')
spk.set_xticklabels(rotation=90);
```
